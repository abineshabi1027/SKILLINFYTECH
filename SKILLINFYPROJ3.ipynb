```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

events = ['Homepage', 'Product Page', 'Add to Cart', 'Checkout', 'Purchase']
data = {
    'UserID': np.random.randint(1000, 2000, 1000),
    'Event': np.random.choice(events, 1000, p=[0.4, 0.3, 0.15, 0.1, 0.05])
}
df = pd.DataFrame(data)

funnel_data = df.groupby('Event')['UserID'].nunique().reindex(events).reset_index()
funnel_data.columns = ['Funnel Stage', 'Unique Users']

funnel_data['Retention Rate (%)'] = (funnel_data['Unique Users'] / funnel_data['Unique Users'].iloc[0]) * 100

print("--- Conversion Funnel Analysis ---")
print(funnel_data)

plt.figure(figsize=(10, 6))
bars = plt.bar(funnel_data['Funnel Stage'], funnel_data['Unique Users'], color='skyblue', edgecolor='black')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom', fontweight='bold')

plt.title('User Conversion Funnel')
plt.xlabel('Pages / Events')
plt.ylabel('Number of Unique Users')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

page_views = df['Event'].value_counts().reset_index()
page_views.columns = ['Page', 'Total Views']
print("\n--- Page Traffic Analysis ---")
print(page_views)
```
